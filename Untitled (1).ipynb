```python
# we are going to visualize for the match of Argentina vs Franc, The World Cup finale
#we need match id for this
MATCH_ID = 3869685
```

```python
match_events_df = sb.events(match_id=MATCH_ID)
```

```python
match_360_df = pd.read_json(f"D:/OneDrive - qqaaqq/Documents/GitHub/open-data/data/three-sixty/{MATCH_ID}.json")
```

```python
match_360_df
```

```python
match_events_df
```

```python
#lets join these 2 data frames
df = pd.merge(left=match_events_df,right=match_360_df,left_on="id",right_on="event_uuid",how="left")
```

```python
df.head(50)
#the FREEZE FRAMES column tells us about who was in the vicinity of the ball
```

```python
df.columns
```

```python
#now lets plot a pass made by messi during the world cup finale
#to find where messi was we need his player id
df.player.unique()
```

```python
MESSI = 5503
df=  df[(df['player_id'] == MESSI) & (df['type'] == "Pass")].reset_index(drop=True)
```

```python
df
```

```python
df.location
#this returns us the location where the ball was x,y coords
# lets convert this to a list of x and y
```

```python
df[['x_start','y_start']] = pd.DataFrame(df.location.to_list(),index=df.index)
```

```python
df
```

```python
#now we need to store the end location of the ball i.e. end coords
df[['x_end','y_end']] = pd.DataFrame(df.pass_end_location.to_list(),index=df.index)
```

```python
df
#now we have the starting point and the ending point of the ball
```

```python
#its time to plot these co-ordinates on the pitch
p = Pitch(pitch_type="statsbomb")
fig, ax = p.draw(figsize=(12,8))


#lets plot messi's pass
df = df[0:1]

p.scatter(x=df['x_start'], y= df['y_start'],ax =ax)
p.lines(xstart = df['x_start'],ystart = df['y_start'],xend = df['x_end'],yend = df['y_end'],comet = True,ax=ax)

#here we dont know the context in which messi is passing.Was it a good pass or a bad?to a teammate or to opponent?no context
#to know the context we'll use the FREEZE FRAME column which tells us about who was near the ball during this pass and will plot them here


for x in df.iloc[0]['freeze_frame']:
    if x['teammate']:
        color = "Blue"
    else:
        color = "red"
    p.scatter(x = x['location'][0],y = x['location'][1],ax =ax ,c = color,s = 100)    
        
```
